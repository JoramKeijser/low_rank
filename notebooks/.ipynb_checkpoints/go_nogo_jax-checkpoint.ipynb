Figure 3 from Mastroguiseppe & Osotjic '18.  Use jax's vmap to vectorize over samples

```python
import numpy as onp
from jax import numpy as np
from jax import random
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette("colorblind")
from jax import vmap, jit
```

```python
# Parameters
N = 1500
g = .8
T = 200
phi = lambda x: np.tanh(x)
key = random.PRNGKey(0)
J = random.normal(key, shape = (N, N)) * g / np.sqrt(N)
```

```python
def sim_network(J, key, I = np.zeros(N,), w = np.zeros(N,)):
    """J: connectivity (N,N)
       I: input (N, )
       w: readout (N, ) 
    """
    key,subkey = random.split(key)
    x = random.normal(subkey, shape=(N, ))
    x_t = []
    print(I.sum())
    for t in range(T):
        if (t> 30) and (t< 50):
            x = 0.9 * x + np.dot(J, phi(x)) + I
        else:
            x = 0.9 * x + np.dot(J, phi(x)) 
        x_t.append(x)
    x_t = np.array(x_t)
    z_t = np.dot(phi(x_t), w)
    return x_t, z_t
```

## Spontaneous activity

```python
x_t, _ = sim_network(J, key=key,I=np.zeros(N))
plt.plot(x_t[:,:5]);
```

## GO-NoGo

```python
keys = random.split(key, 3)
```

```python
# Generate random go / no go patters
keys = random.split(key, 4)
I_go = random.normal(keys[1], shape=(N, ))
I_nogo = random.normal(keys[2], shape=(N, ))
w = random.normal(keys[3], shape=(N, )) * 1/np.sqrt(N) # readout
# Low rank perturbation
n = I_go # left vector correlates with go-stim
m = w # right vector correlates with readout 
J2 = J + .1 * np.dot(m[:,None], n[None,:])
```

```python
def simulate(I, key):
    key, subkey = random.split(key, 2)
    x = random.normal(subkey, shape=(N, )) 
    x_t = []
    for t in range(T):
        if (t>25) and (t<75):
            x = 0.9 * x + np.dot(J2, phi(x)) + I
        else:
            x = 0.9 * x + np.dot(J2, phi(x)) 
        x_t.append(x)
    x_t = np.array(x_t)
    z_t = np.dot(phi(x_t), w)
    return x_t, z_t
```

```python
n_reps = 10
go_resp = []
nogo_resp = []
import time 
t0 = time.time()
for r in range(n_reps):
    key, subkey = random.split(key, 2)
    _, z_go = simulate(key = subkey, I= I_go)
    _, z_nogo = simulate(key = subkey, I = I_nogo)
    go_resp.append(z_go)
    nogo_resp.append(z_nogo)
go_resp = np.array(go_resp)
nogo_resp = np.array(nogo_resp)
t1 = time.time()
print("Loop over reps took %u seconds"%(t1 - t0))
```

```python
plt.plot(go_resp.mean(0))
plt.fill_between(np.arange(T), go_resp.mean(0) - go_resp.std(0), 
                 go_resp.mean(0) + go_resp.std(0), alpha=0.5, label = 'Go')
plt.plot(nogo_resp.mean(0))
plt.fill_between(np.arange(T), nogo_resp.mean(0) - nogo_resp.std(0), 
                 nogo_resp.mean(0) + nogo_resp.std(0), alpha=0.5, label = 'No Go')
plt.legend(fontsize=15)
```

## Vectorization with vmap

First vectorize over samples (random seed), then over samples and input

```python
from jax import vmap
# vectorize over stimuli
batched_simulate = vmap(simulate, (None, 0))
batched_simulate(I_go, key[None])
```

```python
skeys = random.split(key, n_reps+1)
_, go_resp = batched_simulate(I_go, skeys[1:])
```

```python
plt.plot(go_resp.mean(0))
plt.fill_between(np.arange(T), go_resp.mean(0) - go_resp.std(0), 
                 go_resp.mean(0) + go_resp.std(0), alpha=0.5, label = 'Go')
```

```python
# vectorize over inputs and keys
from jax import vmap
batched_simulate = vmap(simulate, (0, 0))
skeys = random.split(key, n_reps+1)
_, go_resp = batched_simulate(np.repeat(I_go[None], n_reps, 0), skeys[1:])
```

```python
# First time: slow
n_reps = 10
t0 = time.time()
skeys = random.split(key, 2*n_reps+1)
_, all_resp = batched_simulate(np.repeat(np.concatenate((I_go[None], I_nogo[None]), 0), n_reps,0),  skeys[1:])
t1 = time.time()
print("Vectorization over reps took %u seconds"%(t1 - t0))
```

```python
go_resp = all_resp[:10]
nogo_resp = all_resp[10:]
plt.plot(go_resp.mean(0))
plt.fill_between(np.arange(T), go_resp.mean(0) - go_resp.std(0), 
                 go_resp.mean(0) + go_resp.std(0), alpha=0.5, label = 'Go')
```

```python
# Second time: quicker
t0 = time.time()
skeys = random.split(skeys[1], 2*n_reps+1) # different seed
_, all_resp = batched_simulate(np.repeat(np.concatenate((I_go[None], I_nogo[None]), 0), n_reps,0),  skeys[1:])
t1 = time.time()
print("Vectorization over reps took %u seconds"%(t1 - t0))
```

```python
go_resp = all_resp[:10]
nogo_resp = all_resp[10:]
plt.plot(go_resp.mean(0))
plt.fill_between(np.arange(T), go_resp.mean(0) - go_resp.std(0), 
                 go_resp.mean(0) + go_resp.std(0), alpha=0.5, label = 'Go')
plt.plot(nogo_resp.mean(0))
plt.fill_between(np.arange(T), nogo_resp.mean(0) - nogo_resp.std(0), 
                 nogo_resp.mean(0) + nogo_resp.std(0), alpha=0.5, label = 'No Go')
plt.legend(fontsize=15)
```

## Analysis

Experimental prediction: positive correlation between correlation of PC1 coefs and connectivity (Figure 3F)

```python
all_rec, all_resp = batched_simulate(np.repeat(np.concatenate((I_go[None], I_nogo[None]), 0), n_reps,0),  skeys[1:])
```

```python
# Select Go trials, during stimulus
X = all_rec[:int(n_reps/2),35:75].mean(0) # 5 * T * N -> T * N
from sklearn.decomposition import PCA
pca= PCA(n_components=1)
X_red = pca.fit_transform(X) #(T *1)
# get coefs on first PC (25) 
coefs = np.dot(X_red.T, X) 
# Correlation between PC 1coefs
C = np.tril(np.dot(coefs.T, coefs),1)
x =C[np.abs(C)>0]
y = J2[np.abs(C)>0]
# random selection for plotting
idx = onp.random.choice(np.arange(x.shape[0]), size=(5000, ))
plt.scatter(x[idx], y[idx])
print(onp.corrcoef(x,y)[0,1])
```

```python
# Select No Go trials 
X = all_rec[int(n_reps/2):, 35:75].mean(0) # 5 * T * N -> T * N
from sklearn.decomposition import PCA
pca= PCA(n_components=1)
X_red = pca.fit_transform(X) #(T *1)
# get coefs on first PC (25) 
coefs = np.dot(X_red.T, X) / N
# Correlation between coefs
C = np.tril(np.dot(coefs.T, coefs),1)
x =C[np.abs(C)>0]
y = J2[np.abs(C)>0]
# random selection for plotting
idx = onp.random.choice(np.arange(x.shape[0]), size=(5000, ))
plt.scatter(x[idx], y[idx])
print(onp.corrcoef(x,y)[0,1])
```
